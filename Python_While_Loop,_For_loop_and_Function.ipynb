# ** While Loop, For loop and Function**

```python
# While loop & Control Statements :(else, break, continue, pass)

# Number Guessing Game

import random

# Generate random number between 1 and 10
secret_number = random.randint(1, 10)
# print(secret_number)

max_attempts = 3

while max_attempts > 0:
    user_input = int(input("Guess the number (between 1 to 10): "))
    max_attempts -= 1

    if user_input < 1 or user_input > 10:
      print(f"Your guess is out of range. Please guess a number between 1 and 10...You only have {max_attempts} attempts left")
      continue
    elif user_input < secret_number:
      print(f"Too Low. Try again...You only have {max_attempts} attempts left")
    elif user_input > secret_number:
      print(f"Too high. Try again....You only have {max_attempts} attempts left")
    elif user_input == secret_number:
      print(f"Congratulations! You guessed the correct number.")
      break

else:
      print(f"Better luck next time!")
```

```python
# For loop: Multiplication Table Generator

user_input = int(input("Enter a number to generate a multiplication table: "))
i = 1
for i in range(1,11):
  result = user_input*i
  print(f"{user_input} * {i} = {result}")
```

```python
# Function: BMI Calculator

def calculate_bmi(weight, height):
    bmi = round(weight / (height ** 2),2)
    return bmi

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = calculate_bmi(weight, height)
print(f"Your BMI is: {bmi}")
```
